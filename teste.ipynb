```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Define o tamanho visual dos eixos
eixo_maximo = 50

# Define o tamanho do raio do setor circular
raio = 0

# Define o centro do setor na origem (0, 0)
center = (0, 0)

# Cria uma figura e um eixo
fig, ax = plt.subplots(figsize=(8, 8))

# Cria um setor circular (cunha) centrado na origem, com o raio definido
wedge = patches.Wedge(
    center,
    raio,
    0,
    90,
    fill=False,
    color='b',
    linestyle='-',
    linewidth=2
)

# Adiciona o setor ao eixo
ax.add_patch(wedge)

# Desenha linhas do centro até as bordas para representar os raios
ax.plot([center[0], center[0] + raio], [center[1], center[1]], 'r--', lw=1)
ax.plot([center[0], center[0]], [center[1], center[1] + raio], 'r--', lw=1)

# Adiciona uma anotação de texto para o valor do raio
ax.text(center[0] + raio / 2, center[1], f'Raio: {raio}', ha='center', va='bottom', color='green')

# Define os limites dos eixos para o tamanho visual desejado
#ax.set_xlim([0, eixo_maximo])
#ax.set_ylim([0, eixo_maximo])
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_title(f'Setor Circular (Raio: {raio}) com Eixos Visuais de {eixo_maximo}')
plt.yticks(range(0,50,10))
plt.xticks(range(0,50,10)) 

# Garante que a proporção da imagem seja mantida
plt.axis('equal')

# Adiciona linhas de grade para melhor visualização
plt.grid(True)

# Salva o gráfico
plt.savefig('circular_sector_30_on_50_axes.png')
```
