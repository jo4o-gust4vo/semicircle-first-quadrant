```python
import matplotlib.pyplot as plt
```

```python
circle = plt.Circle((20, 10), 10, fill=False)

# Create a figure and axis
fig, ax = plt.subplots()

# Add the circle to the axis
ax.add_patch(circle)

# Set axis limits and show the plot
ax.set_xlim(0,11)
ax.set_ylim(0,11)
plt.axis('equal')  # To ensure the aspect ratio is maintained



plt.show()

plt.close()
```

```python


# Create the circle
circle = plt.Circle((20, 10), 10, fill=False)

# Create a figure and axis
fig, ax = plt.subplots()

# Add the circle to the axis
ax.add_patch(circle)

# Get the center and radius of the circle
center = circle.center
radius = circle.radius

# Draw a line from the center to the edge to represent the radius
ax.plot([center[0], center[0] + radius], [center[1], center[1]], 'r--', lw=2)

# Add a text annotation for the radius value
ax.text(center[0] + radius/2, center[1], f'Raio: {radius}', ha='center', va='bottom', color='green')

# Set axis limits and show the plot
ax.set_xlim(0, 20)
ax.set_ylim(0, 20)
plt.axis('equal')  # To ensure the aspect ratio is maintained
plt.savefig(r'D:\Documentos\Estudos\Faculdade\Sétimo semestre\COMPUTAÇÃO GRÁFICA\static\imagem\circulo.png')
plt.show()
```
